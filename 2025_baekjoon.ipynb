```python
A = list(map(int, input().split()))
k = int(input())

def partition(A, p, r):
    pivot = A[r]
    L = p
    R = r - 1
    while L <= R:
        while L <= R and A[L] <= pivot:
            L += 1
        while L <= R and A[R] > pivot:
            R -= 1
        if L < R:
            A[L], A[R] = A[R], A[L]
    A[L], A[r] = A[r], A[L]
    return L

def select(A, p, r, k):
    if p == r:
        return A[p]
    q = partition(A, p, r)
    m = q - p + 1  # 왼쪽 부분의 원소 개수 (pivot 포함)
    if k == m:
        return A[q]
    elif k < m:
        return select(A, p, q - 1, k)
    else:
        return select(A, q + 1, r, k - m)

ans = select(A, 0, len(A) - 1, k)
print(ans)
```

```python
import math
import heapq

# 프림 알고리즘: 추가되는 노드 포함한 방문 노드 전체에서 갈 수 있는 간선의 가중치 중 최소 간선 그리디
N = int(input())
nodes = [list(map(float, input().split())) for _ in range(N)]

def distance(star1, star2):
    x1, y1 = star1
    x2, y2 = star2
    w = math.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
    return w

# 인접 리스트로 그래프 구성
graph = [[] for _ in range(N)]
for i in range(N):
    for j in range(i + 1, N):
        w = distance(nodes[i], nodes[j])
        graph[i].append((j, w))
        graph[j].append((i, w))

pq = []  # 우선순위 큐(힙)
mst = []  # 선택한 간선 저장 배열
visited = [False] * N
visited[0] = True
total_weight = 0  # 최소 신장 트리의 총 가중치

for v, w in graph[0]:#0번 노드와 연결된 정보 큐에 추가
    heapq.heappush(pq, (w, 0, v))#가중치,시작노드.도착노드
while pq:
    w, prev, u = heapq.heappop(pq)#가중치,prev:이전방문노드,u:방문할노드 힙에서 꺼냄
    if not visited[u]:
        visited[u] = True
        mst.append((prev, u, w))
        total_weight += w
        if len(mst) == N - 1:
            break
        for v, w2 in graph[u]:#u에서 갈수 있는 노드 탐색
            if not visited[v]:
                heapq.heappush(pq, (w2, u, v))

print(total_weight)
```

```python
# 1976 여행가자
N = int(input())  # 전체 도시 수
M = int(input())  # 여행 계획에 속한 도시들의 수

class UnionFind:
    def __init__(self, size):
        self.id = list(range(size))  # size가 5면 [0,1,2,3,4]
        self.rank = [0] * size  # rank 초기화

    def find(self, x):  # x의 루트 원소를 찾음
        if self.id[x] != x:  #self.id:자기자신의 id, self.id[x]:x의 부모노드
            self.id[x] = self.find(self.id[x])  #부모노드가 자기자신,즉 루트노드가 아니면 find 재귀로 수행(경로압축)
        return self.id[x]

    def unite(self, a, b):
        a, b = self.find(a), self.find(b)  # 각각의 루트 찾기
        if a == b:  # 이미 같은 집합
            return
        if self.rank[a] > self.rank[b]:  # a의 깊이가 더 크면
            self.id[b] = a  #
        elif self.rank[a] < self.rank[b]:
            self.id[a] = b
        else:  # 깊이가 같으면
            self.id[b] = a  # b를 a에 합침
            self.rank[a] += 1  # a의 깊이 증가

#1번부터 N번까지 사용하므로 N+1
uf = UnionFind(N + 1)#객체생성,uf변수에 저장

for i in range(N):
    A = list(map(int, input().split()))
    for j in range(N):
        if A[j] == 1:
            uf.unite(i + 1, j + 1)
# 여행 계획 입력
plan = list(map(int, input().split()))

possible = True
root = uf.find(plan[0])  #=uf.find(1)
for city in plan:
    if uf.find(city) != root:  # 같은 집합에 속하지 않으면
        possible = False
        break

if possible:
    print("YES")
else:
    print("NO")
```

```python
#2493 스택에는 값이 아닌 인덱스를 저장
from collections import deque
n=int(input())
A=list(map(int,input().split()))
stack=deque()
b=[0]*n#스택에 있는 요소 인덱스만 저장하는 리스트

for i in range(0,n):
  while stack and A[i]>A[stack[-1]]:  #스택에 값이 있고 현재탑이 스택에 있는 요소보다 크면
     stack.pop()#스택에 있는 값을 지운다
  if stack:#스택이 값이 있으면
     b[i]=stack[-1]+1#해당 인덱스(번째니까 +1)를 b에 추가
  stack.append(i)  #스택에 값이 없으면 stack에 해당 탑의 인덱스 추가

# 한 줄로 정수만 출력
print(' '.join(map(str, b)))
```

```python
from collections import deque
n=int(input())
A=list(map(int,input().split()))
stack=deque()
stack.append(A[0])
b=[]
b.append(0)

for i in range(1,n):  # n-1이 아니라 n으로 수정
  if stack and A[i]>A[i-1]:  # 이전값보다 크면
     stack.pop()
     stack.append(A[i])
     b.append(i)  # i+1이 아니라 i로 수정 (인덱스는 0부터 시작)
  else:  # A[i]<=A[i-1] 이전값이 크면 b에 이미 저장된 인덱스 출력
     b.append(b[-1]+1)

# 한 줄로 정수만 출력
print(' '.join(map(str, b)))
```
